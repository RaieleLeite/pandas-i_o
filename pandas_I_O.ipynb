## Instituiçao financeira - Banco de dados local

```python
import pandas as pd
import sqlalchemy

from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')
# Create_engine() é uma função que configura a conexão com um banco de dados.
# Sqlalchemy 'sqlite:///:memory:' é a URL do banco de dados.
```

```python
url_banco = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
dados_banco = pd.read_csv(url_banco)
dados_banco.head()
```

```python
# Transforma o DataFrame para o formato do banco de dados
dados_banco.to_sql('clientes', engine, index=False)
```

```python
# Inspect: Inspeciona o banco de dados
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
empregados.head()
```

```python
empregados.to_sql('empregados', con=engine, index=False)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=['ID_Cliente','Grau_escolaridade','Rendimento_anual'])
```

```python
query_clientes = 'SELECT * from clientes'
pd.read_sql(query_clientes, engine)
```

```python
query_delete = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    result = conn.execute(text(query_delete))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query_update = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    result = conn.execute(text(query_update))
    conn.commit()
```

```python
pd.read_sql_table('clientes', engine)
```
